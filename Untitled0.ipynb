```python
# @title Default title text
from sentence_transformers import SentenceTransformer
import faiss
import gradio as gr
import numpy as np
import pandas as pd

data = {
    "questions": [
        "What is AI?",
        "How does machine learning work?",
        "What is a neural network?",
        "Explain deep learning.",
        "What is NLP?",
        "Hi there! how can I help you today?"
        ],
    "answers": [
        "AI is the simulation of human intelligence in machines.",
        "Machine learning allows systems to learn from data without being explicitly programmed.",
        "A neural network is a series of algorithms that mimic the operations of a human brain.",
        "Deep learning is a subset of ML using layered neural networks.",
        "NLP stands for Natural Language Processing, a field focused on language interaction between computers and humans.",
        "I am fine! thanks. I need help in learning AI"
    ]
}
df = pd.DataFrame(data)

model = SentenceTransformer('all-MiniLM-L6-v2')  # Fast and free

# Generate embeddings
question_embeddings = model.encode(df['questions'].tolist())

# Build FAISS index
index = faiss.IndexFlatL2(question_embeddings.shape[1])
index.add(np.array(question_embeddings))

def answer_question(user_input):
    query_embedding = model.encode([user_input])
    D, I = index.search(np.array(query_embedding), k=1)
    best_answer = df.iloc[I[0][0]]['answers']
    return best_answer

iface = gr.Interface(fn=answer_question,
                     inputs="text",
                     outputs="text",
                     title="Simple AI FAQ Bot",
                     description="Ask a question from the AI FAQ database.")

iface.launch()

```

```python
!pip install -q transformers sentence-transformers faiss-cpu gradio


from sentence_transformers import SentenceTransformer
import faiss
import gradio as gr
import numpy as np
import pandas as pd

data = {
    "questions": [
        "What is AI?",
        "How does machine learning work?",
        "What is a neural network?",
        "Explain deep learning.",
        "What is NLP?"
    ],
    "answers": [
        "AI is the simulation of human intelligence in machines.",
        "Machine learning allows systems to learn from data without being explicitly programmed.",
        "A neural network is a series of algorithms that mimic the operations of a human brain.",
        "Deep learning is a subset of ML using layered neural networks.",
        "NLP stands for Natural Language Processing, a field focused on language interaction between computers and humans."
    ]
}
df = pd.DataFrame(data)


model = SentenceTransformer('all-MiniLM-L6-v2')
question_embeddings = model.encode(df['questions'].tolist())
index = faiss.IndexFlatL2(question_embeddings.shape[1])
index.add(np.array(question_embeddings))

chat_history = []

def answer_with_memory(user_input):
    # Save user's message
    chat_history.append(("User", user_input))

    # Search for best match
    query_embedding = model.encode([user_input])
    D, I = index.search(np.array(query_embedding), k=1)
    best_answer = df.iloc[I[0][0]]['answers']

    # Save assistant's reply
    chat_history.append(("Assistant", best_answer))

    # Build conversation string
    conversation = ""
    for role, message in chat_history:
        conversation += f"{role}: {message}\n"

    return conversation


iface = gr.Interface(
    fn=answer_with_memory,
    inputs="text",
    outputs="textbox",
    title="Q&A Assistant with Memory",
    description="Ask anything about AI. The assistant remembers your past questions."
)

iface.launch()
```

```python

# SETUP: Install Required Packages
!pip install -q transformers sentence-transformers faiss-cpu accelerate bitsandbytes gradio PyMuPDF python-docx beautifulsoup4 requests

# IMPORT LIBRARIES
from transformers import AutoTokenizer, AutoModelForCausalLM
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
import fitz  # PyMuPDF
from docx import Document
import requests
from bs4 import BeautifulSoup
import gradio as gr

from huggingface_hub import login
login()  # paste your token when prompted

"""
from google.colab import files
uploaded = files.upload()
file_path = list(uploaded.keys())[0]
"""
# TEXT EXTRACTION FUNCTIONS
def extract_text_from_pdf(file_path):
    doc = fitz.open(file_path)
    return "".join(page.get_text() for page in doc)

def extract_text_from_docx(file_path):
    doc = Document(file_path)
    return "\n".join(para.text for para in doc.paragraphs)

def extract_text_from_url(url):
    try:
        headers = {"User-Agent": "Mozilla/5.0"}  # avoid 403
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()

        soup = BeautifulSoup(response.text, 'html.parser')
        text = soup.get_text(separator=" ", strip=True)
        if not text or len(text) < 100:
            return "Failed to extract readable content. Try another URL."
        return text
    except Exception as e:
        return f"Error fetching URL: {str(e)}"

"""
def extract_text_from_url(url):
    try:
        page = requests.get(url)
        soup = BeautifulSoup(page.content, 'html.parser')
        return soup.get_text()
    except:
        return "Failed to fetch content from URL."
"""
#  TEXT CHUNKING FUNCTION
def chunk_text(text, max_words=100, overlap=20):
    words = text.split()
    return [" ".join(words[i:i+max_words]) for i in range(0, len(words), max_words - overlap)]

#  LOAD EMBEDDER & INIT INDEX
embedder = SentenceTransformer('all-MiniLM-L6-v2')
index = faiss.IndexFlatL2(384)  # 384 = MiniLM vector size
doc_chunks = []

#  LOAD TINYLLAMA MODEL
model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto", )

#  MEMORY
chat_history = []

#  LOAD & INDEX DOCUMENTS FUNCTION
def load_documents(file_objs=[], urls=[]):
    global doc_chunks, index
    all_texts = []

    for file in file_objs:
        if file.name.endswith(".pdf"):
            all_texts.append(extract_text_from_pdf(file.name))
        elif file.name.endswith(".docx"):
            all_texts.append(extract_text_from_docx(file.name))

    for url in urls:
        all_texts.append(extract_text_from_url(url))

    combined_text = "\n".join(all_texts)
    doc_chunks = chunk_text(combined_text)
    embeddings = embedder.encode(doc_chunks)
    index = faiss.IndexFlatL2(embeddings.shape[1])
    index.add(embeddings)

    return f"Loaded {len(doc_chunks)} document chunks from files and URLs."

#  Q&A FUNCTION WITH MEMORY + CONTEXT
def generate_response(user_input):
    global chat_history, doc_chunks

    # Retrieve context from FAISS
    query_embedding = embedder.encode([user_input])
    _, I = index.search(np.array(query_embedding), k=3)
    context = doc_chunks[I[0][0]] if len(doc_chunks) > 0 else ""

    # Build chat history
    history_prompt = ""
    for user_msg, bot_msg in chat_history[-3:]:
        history_prompt += f"<|user|>{user_msg}</s>\n<|assistant|>{bot_msg}</s>\n"

    # Prompt for TinyLlama
    prompt = f"<|system|>You are a helpful assistant using document and website context below.\n\nContext:\n{context}\n\n{history_prompt}<|user|>{user_input}</s>\n<|assistant|>"

    input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to(model.device)
    output = model.generate(input_ids, max_new_tokens=256)
    response = tokenizer.decode(output[0], skip_special_tokens=True)

    # Clean up and extract assistant response
    reply = response.split("<|assistant|>")[-1].strip()
    chat_history.append((user_input, reply))

    return reply

#  GRADIO INTERFACE
def chat_interface(user_input, url_input, files):
    if url_input or files:
        load_documents(files, urls=[url.strip() for url in url_input.split(",") if url.strip()])
        return generate_response(user_input)
    else:
        return generate_response(user_input)

gr.Interface(
    fn=chat_interface,
    inputs=[
        gr.Textbox(label="Your Question"),
        gr.Textbox(label="Optional: Website URLs (comma-separated)"),
        gr.File(label="Optional: Upload PDF/DOCX", file_types=[".pdf", ".docx"], file_count="multiple"),
    ],
    outputs="text",
    title="🧠 TinyLlama Chatbot with Multi-Doc + Web Context + Memory",
    description="Ask questions with context from your uploaded documents and websites. Powered by TinyLlama running locally in Colab."
).launch()
```
